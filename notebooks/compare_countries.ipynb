# Cross-Country Solar Potential Comparison

This notebook analyzes and compares solar potential metrics (GHI, DNI, DHI) across Benin, Sierra Leone, and Togo.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

sns.set_theme() 
sns.set_palette("husl")
```

```python
# Load the cleaned datasets
togo_df = pd.read_csv('data/togo_clean.csv')
sierraleone_df = pd.read_csv('data/sierraleone_clean.csv')

# Add country column to each dataset
togo_df['country'] = 'Togo'
sierraleone_df['country'] = 'Sierra Leone'

# Combine datasets
combined_df = pd.concat([togo_df, sierraleone_df], ignore_index=True)
```

## Metric Comparisons - Boxplots

```python
# Create boxplots for GHI, DNI, DHI
metrics = ['GHI', 'DNI', 'DHI']

for metric in metrics:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=combined_df, x='country', y=metric)
    plt.title(f'{metric} Distribution by Country')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

## Summary Statistics Table

```python
# Create summary statistics table
summary_stats = []
for metric in metrics:
    metric_stats = combined_df.groupby('country')[metric].agg(['mean', 'median', 'std']).round(2)
    metric_stats['metric'] = metric
    summary_stats.append(metric_stats)

summary_table = pd.concat(summary_stats, axis=0)
print("\nSummary Statistics:")
print(summary_table)
```

## Statistical Testing

```python
# Statistical Testing - One-way ANOVA for each metric
print("\nStatistical Tests:")
for metric in metrics:
    togo_values = togo_df[metric]
    sierraleone_values = sierraleone_df[metric]
    
    # Perform one-way ANOVA
    f_stat, p_value = stats.f_oneway(togo_values, sierraleone_values)
    print(f"\n{metric}:")
    print(f"One-way ANOVA p-value: {p_value:.4f}")
```

```python
# Bar chart of average GHI by country
plt.figure(figsize=(8, 6))
sns.barplot(data=combined_df, x='country', y='GHI', ci='sd')
plt.title('Average GHI by Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
